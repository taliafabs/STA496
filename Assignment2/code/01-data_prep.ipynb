```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# setup for categorical variables
gender_categories = pd.CategoricalDtype(categories=['Male', 'Female'], ordered=True)
age_categories = pd.CategoricalDtype(categories=['18–24', '25–34', '35–44', '45–54', '55–64', '65–74', '75+'], ordered=True)
race_map = {
    'White': 'White',
    'Black': 'Black',
    'Hispanic': 'Hispanic',
    'Asian': 'All Other Races',
    'Two or more races': 'All Other Races',
    'Other': 'All Other Races',
    'Native American': 'All Other Races',
    'Middle Eastern': 'All Other Races'
}
```

```python
# select relevant columns
# region and urbancity not available in 2016 dataset
ces16_subset = ces16_raw[[
    'CC16_364b', # presidential vote
    'CC16_364c', # preferred presidential candidate
    'race', # self reported race
    'birthyr', # year respondent was born
    'gender', # gender identity
    'educ', # highest level of educational attainment
    'inputstate', # state that respondent lives in
    'CC16_302', # national economy over the past year
    'CC16_303', # household income past 4yrs
    'CC16_320a', # obama (incumbent pres) approval
    'tookpost', # whether respondent participated in post election wave
    'commonweight',  # weight of respondent common,
    'commonpostweight' # weight of the respondent if they took post
]]

# recode race
ces16_subset['race'] = ces16_subset['race'].map(race_map)

# create presvore16 column
```

```python
# 2020
!wget --no-check-certificate "https://dataverse.harvard.edu/api/access/datafile/4949545" -O CES20_Common_OUTPUT_vv.dta
ces20_raw = pd.read_stata("CES20_Common_OUTPUT_vv.dta")
```

```python
ces20_raw.head()
```

```python
ces20_subset = ces20_raw[['CC20_410', # 2020 presidential vote
                          'race', # self identified race
                          'birthyr', # year respondent was born
                          'gender', # gender identity
                          'educ', # highest level of educational attainment
                          'inputstate', # state that respondent lives in
                          'region', # census region that respondent lives in
                          'urbancity',
                          'CC20_302', # national economy over the past year
                          'CC20_303', # household income past year
                          'CC20_320a', # trump approval
                          'tookpost', # whether respondent participated in post election wave
                          'commonweight',  # weight of respondent common,
                          'commonpostweight' # weight of the respondent if they took post
                          ]]

ces20_subset['age'] = 2020 -  ces20_subset['birthyr']

# create age bracket variable
ces20_subset['age_bracket'] = pd.cut(
    ces20_subset['age'],
    bins=[17, 24, 34, 44, 54, 64, 74, 100],
    labels=[
        '18–24', '25–34', '35–44', '45–54', '55–64', '65–74', '75+'
    ]
)

# only include trump, biden, and third party categories
# ces20_subset = ces20_subset[
#     (ces20_subset['CC24_364a'] == "Kamala Harris (Democrat)") |
#     (ces20_subset['CC24_364a'] == "Donald Trump (Republican)") |
#     (ces20_subset['CC24_364b'] == "Kamala Harris (Democrat)") |
#     (ces20_subset['CC24_364b'] == "Donald Trump (Republican)") |
#     (ces20_subset['CC24_364a'] == "Someone else") |
#     (ces20_subset['CC24_364b'] == "Other")
# ]

# recode race
ces20_subset['race'] = ces20_subset['race'].map(race_map)
ces20_subset = ces20_subset.rename(columns={'CC20_410': 'presvote20',
                                            'CC20_302': 'econ_past_year',
                                            'CC20_303': 'family_income_past_year',
                                            'CC20_320a': 'incumbent_pres_approval',
                                            'inputstate': 'state'
                                            })

# only include trump, biden, and other
trump_voters = ces20_subset['presvote20'] == "Donald J. Trump (Republican)"
biden_voters = ces20_subset['presvote20'] == "Joe Biden (Democrat)"
other_voters = ces20_subset['presvote20'] == "Other"
ces20_subset = ces20_subset[trump_voters | biden_voters | other_voters]

#
# remove all na's
ces20_subset = ces20_subset.dropna()

# make predeictors categorical
ces20_subset['presvote20'] = ces20_subset['presvote20'].astype('category')
ces20_subset['age_bracket'] = ces20_subset['age_bracket'].astype(age_categories)
ces20_subset['gender'] = ces20_subset['gender'].astype(gender_categories)
ces20_subset['educ'] = ces20_subset['educ'].astype('category')
ces20_subset['state'] = ces20_subset['state'].astype('category')
ces20_subset['region'] = ces20_subset['region'].astype('category')
ces20_subset['urbancity'] = ces20_subset['urbancity'].astype('category')
ces20_subset['family_income_past_year'] = ces20_subset['family_income_past_year'].astype('category')
ces20_subset['econ_past_year'] = ces20_subset['econ_past_year'].astype('category')
ces20_subset['incumbent_pres_approval'] = ces20_subset['incumbent_pres_approval'].astype('category')
ces20_subset['race'] = ces20_subset['race'].astype('category')

ces20_subset = ces20_subset[ces20_subset['tookpost']=="Yes"]
ces20_subset = ces20_subset[ces20_subset['gender'].isin(["Male", "Female"])]
```

```python
ces20_subset['presvote20'].value_counts()
```

```python
ces20_clean = ces20_subset[['presvote20',
                            'age_bracket',
                            'gender',
                            'race',
                            'educ',
                            'state',
                            'region',
                            'urbancity',
                            'econ_past_year',
                            'family_income_past_year',
                            'incumbent_pres_approval',
                            'tookpost',
                            'commonweight',
                            'commonpostweight'
                            ]]
```

```python
# create dummy variables
def create_dummy_interaction(col1, col2):
  result = pd.DataFrame()
  col1_dummies = pd.get_dummies(col1, drop_first=True)
  col2_dummies = pd.get_dummies(col2, drop_first=True)
  for col1_val in col1_dummies.columns:
    for col2_val in col2_dummies.columns:
      result[f'{col1_val}_{col2_val}'] = col1_dummies[col1_val] * col2_dummies[col2_val]
  return result

# make it only include trump and biden voters


ces20_clean = ces20_clean[(ces20_clean['presvote20'] == 'Donald J. Trump (Republican)') | (ces20_clean['presvote20'] == 'Joe Biden (Democrat)')]
ces20_clean['vote_incumbent'] = np.where(ces20_clean['presvote20'] == 'Donald J. Trump (Republican)', 1, 0)
ces20_clean = ces20_clean[ces20_clean['tookpost'] == "Yes"]
ces20_clean = ces20_clean[['vote_incumbent',
                           'age_bracket',
                           'race',
                           'gender',
                           'educ',
                           'state',
                           'region',
                           'urbancity',
                           'econ_past_year',
                           'family_income_past_year',
                           'incumbent_pres_approval',
                           ]]
```

```python
race_gender_interaction_terms = create_dummy_interaction(ces20_clean['race'], ces20_clean['gender'])
race_educ_interaction_terms = create_dummy_interaction(ces20_clean['race'], ces20_clean['educ'])
race_region_interaction_terms = create_dummy_interaction(ces20_clean['race'], ces20_clean['region'])
race_urban_interaction_terms = create_dummy_interaction(ces20_clean['race'], ces20_clean['urbancity'])

ces20_analysis_dummies = pd.concat([
    pd.get_dummies(ces20_clean, drop_first=True),
    race_gender_interaction_terms,
    race_educ_interaction_terms,
    race_region_interaction_terms,
    race_urban_interaction_terms], axis=1)
```

```python
# 2024
url = "https://github.com/taliafabs/STA496/raw/main/MidtermPaper/data/ces24_analysis_data.parquet"
ces24_df = pd.read_parquet(url, engine="pyarrow")
race_map = {
    'White': 'White',
    'Black': 'Black',
    'Hispanic': 'Hispanic',
    'Asian': 'All Other Races',
    'Two or more races': 'All Other Races',
    'Other': 'All Other Races',
    'Native American': 'All Other Races',
    'Middle Eastern': 'All Other Races'
}
ces24_df['race'] = ces24_df['race'].map(race_map)
ces24_df = ces24_df[(ces24_df['gender'] == "Woman") | (ces24_df['gender'] == "Man")]

gender_map = {
    "Woman": "Female",
    "Man": "Male"
}
ces24_df['gender'] = ces24_df['gender'].map(gender_map)

# filtering
ces24_df = ces24_df[ces24_df['tookpost'] == "Yes"]
ces24_df['vote_incumbent'] = 1 - ces24_df['vote_trump']
ces24_df = ces24_df.rename(columns={'biden_approval': 'incumbent_pres_approval'})
ces24_df = ces24_df[['vote_incumbent',
                    #  'vote_trump',
                           'age_bracket',
                           'race',
                           'gender',
                           'educ',
                           'state',
                           'region',
                           'urbancity',
                           'econ_past_year',
                           'family_income_past_year',
                           'incumbent_pres_approval',
                           ]]

# make things categorical
ces24_df['race'] = ces24_df['race'].astype('category')
ces24_df['gender'] = ces24_df['gender'].astype(gender_categories)
ces24_df['educ'] = ces24_df['educ'].astype('category')
ces24_df['age_bracket'] = ces24_df['age_bracket'].astype(age_categories)

race_gender_interaction_terms = create_dummy_interaction(ces24_df['race'], ces24_df['gender'])
race_educ_interaction_terms = create_dummy_interaction(ces24_df['race'], ces24_df['educ'])
race_region_interaction_terms = create_dummy_interaction(ces24_df['race'], ces24_df['region'])
race_urban_interaction_terms = create_dummy_interaction(ces24_df['race'], ces24_df['urbancity'])

ces24_dummies = pd.concat([
    pd.get_dummies(ces24_df, drop_first=True),
    race_gender_interaction_terms,
    race_educ_interaction_terms,
    race_region_interaction_terms,
    race_urban_interaction_terms,
], axis=1)

# make it binary (remove third party voters for modeling part)
# ces24_dummies[ces24_dummies['presvote24_Other'] == 0]

# drop the presvote24 related columns
# ces24_dummies = ces24_dummies.drop(columns=['presvote24_Other'])
# ces24_dummies = ces24_dummies.drop(columns=['presvote24_Kamala Harris (D)'])
# ces24_dummies = ces24_dummies.drop(columns=['birthyr'])
# ces24_dummies = ces24_dummies.drop(columns=['commonweight'])
# ces24_dummies = ces24_dummies.drop(columns=['commonpostweight'])
ces24_dummies
```
